```python
print("Size of the Data Set is : ",len(data))
# Separate features (redness and weight) and labels (fruit type)
x = data[:,:2]
y = data[:,2]
```

```python

# Split dataset into training and testing sets
x_main, x_test, y_main, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
x_val,x_train,y_val,y_train = train_test_split(x_main,y_main,test_size=0.8)
```

```python
# Create a KNN classifier with k=1
k = 1
while k<=7:
        print("Value of k is - ",k)
        knn_classifier = KNeighborsClassifier(n_neighbors=k)
# Train the classifier on the training data
        knn_classifier.fit(x_train, y_train)
        print("\nX_test : \n\n",x_test)
#Train the classifier on the training data
        knn_classifier.fit(x_train, y_train)  
        print("\nx_val : \n\n",x_val)
# Predict the fruit types for the test data
        y_pred=knn_classifier.predict(x_test)
        print("\ny_pred : \n\n",y_pred)
#Predict the fruit types for the validation data
        yval_pred= knn_classifier.predict(x_val)
        print("\nyval_pred : \n\n",yval_pred)
# Calculate the accuracy of the model
        validation_accuracy =accuracy_score(y_val,yval_pred)
        test_accuracy = accuracy_score(y_test, y_pred)
        print("\nAccuracy of test\n", test_accuracy)
        print("\nAccuracy of validation\n",validation_accuracy)
# Confusion matrix to check the Performance of Algorithm
        cm = confusion_matrix(y_test,y_pred)
        plt.figure(figsize=(5,2))
        sn.heatmap(cm,annot=True)
        plt.xlabel('Predicted')
        plt.ylabel('Truth')
        k=k+2
        
        df = pd.DataFrame(data, columns=['Redness', 'Weight', 'Fruit Type'])

        # Plot the data samples
        plt.figure(figsize=(8, 6))
        for fruit_type in df['Fruit Type'].unique():
            subset = df[df['Fruit Type'] == fruit_type]
            plt.scatter(subset['Redness'], subset['Weight'], label=fruit_type)

        plt.xlabel('Redness Value')
        plt.ylabel('Weight (grams)')
        plt.title('Apple vs Orange Classification')
        plt.legend()
        plt.show()
```
